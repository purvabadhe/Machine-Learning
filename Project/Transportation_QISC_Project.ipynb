```python
from haversine import haversine
import pandas as pd
import numpy as np
import folium
```

```python
supplier = pd.read_excel("Transportation.xlsx",sheet_name=0)
customer = pd.read_excel("Transportation.xlsx",sheet_name=1)
```

```python
center = supplier[['lat','long']].mean().values
```

```python
m = folium.Map(location=center,zoom_start=13)

for index,row in supplier.iterrows():
    folium.Marker(location=[row[1],row[2]],popup=f"Branch {row[0]}",icon=folium.Icon(color="red", icon="home")).add_to(m)

for index,row in customer.iterrows():
    folium.Marker(location=[row[1],row[2]],popup=f"Name: {row[0]}",icon=folium.Icon(color="blue", icon="home")).add_to(m)
```

```python
m
```

```python
cost = dict()
for index1,data1 in supplier.iterrows():
    cost[f'{data1[0]}'] = dict()
    for index2,data2 in customer.iterrows():
        distance = haversine((data1[1],data1[2]),(data2[1],data2[2]))
        cost[f'{data1[0]}'][f'{data2[0]}'] = round(distance,2)
```

```python
cost_matrix = []
for index1,data1 in supplier.iterrows():
    cost_vector = []
    for index2,data2 in customer.iterrows():
        distance = haversine((data1[1],data1[2]),(data2[1],data2[2]))
        cost_vector.append(round(distance,2))
    cost_matrix.append(cost_vector)
```

```python
distance_matrix = pd.DataFrame(cost_matrix,index=supplier['Branch'].values,columns=customer['Name'].values)
distance_matrix
```
